```python
# Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
# to get the basic info

df.info()
```

```python
# to drop irrelavent columns

df = df.drop(['customerID'],axis=1)
df
```

```python
# to convert object into Integer

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
# to plot boxplot and Displot for outliers(univariant)

for i in Numeric_Data:
    print("||",i,"||")
    sns.boxplot(df[i])
    sns.displot(df[i])
    plt.show()
    print('='*50)
```

```python
# count plot

for i in df.drop(['tenure','MonthlyCharges','TotalCharges'],axis=1):
    print('||',i,'||')
    sns.countplot(x=df[i])
    plt.show()
```

```python
# to create histplot | kdeplot

for i in df[['MonthlyCharges','TotalCharges']]:
    print("||",i,"||")
    sns.histplot(df[i],kde=True)
    plt.show()
```

```python
# to create a scatterplot to analyse the Total charge and monthly charge (BIvariant)

sns.scatterplot(data=df,x='TotalCharges',y='MonthlyCharges',hue='Contract')
plt.show()
```

```python
# to split the data

x=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
# standard scaling

from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x = ss.fit_transform(x)
```

```python
# to split data into train and test

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# import and load Logistic Regression

from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

```python
df.isnull().sum()
```

```python
# predict

y_pred_train = lr.predict(x_train)
y_pred_test = lr.predict(x_test)
```

```python
# classification report

from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred_train,y_train))
```

```python
print(classification_report(y_pred_test,y_test))
```
